### 問1

```

**
***

*****
```

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
size = 5
for i in range(size):
    print(' ' * (size - i - 1) + '*')
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
size = 5
mid = size // 2
for i in range(size):
    # 上半分は i が増えるごとに先頭スペースが増え、下半分は減る
    spaces_before = i if i <= mid else (size - 1 - i)
    # 先頭と末尾の '*' の間のスペース数
    spaces_between = size - 2 - 2 * spaces_before
    if spaces_between >= 0:
        print(' ' * spaces_before + '*' + ' ' * spaces_between + '*')
    else:
        print(' ' * spaces_before + '*')
````

### 問5

以下の図形を出力してください。

```
    *
   ***
  *****
 *******
*********
```

````python
size = 5
for i in range(size):
    print(' ' * (size - i - 1) + '*' * (2 * i + 1))
````
